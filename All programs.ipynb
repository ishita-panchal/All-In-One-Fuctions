```python
def budget_game():
    #budget game
    print("WELCOME TO THE 2023 IPL TICKETS BOOKINGS")
    #greeting user
    name=input("\nEnter your name: ")
    print("Hello", name.title())

    #taking input from the user 
    budget=int(input("\nEnter your budget: "))
    print()
    #putting condtions according to the cusotmers budget 
    if budget<400:
        print('Sorry sir/mam you dont have enough budget to buy the IPL tickets.')
        print('You should have a minimum budget of 400rs.')
    #for block C1&D1....
    elif budget>=400 and budget<500:
        print("Your ticket has been booked for the block C1&D1.")
        print("By adding some money to your budget you can buy tickets for the block B1,D,E&F1.") 
    #for block B1,D,E&F1.....
    elif budget>=500 and budget<900:  
        print("Your ticket has been booked for the block B1,D,E & F1. ")
    #for block F
    elif budget>=900 and budget<1000:
        print("Your ticket has been booked for the block F.")
    #for block C&K
    elif budget>=1000 and budget<1800:
        print("Your ticket has been booked for the block C&K.")
    #for block Clubhouse lower
    elif budget>=1800 and budget<2100:
        print("Your ticket has been booked for the block Clubhouse Lower.")
    #for block clubhouse upper
    elif budget>=2100 and budget<9000:
        print("Your ticket has been booked for the block Clubhouse Upper.")
    #else passs
    else: 
        print('No tickets are available within this range....')
    print()
    print("Thankyou!!!!")


#defining function
def cube_game():
    #welcome
    print('Welcome to the Cube Game!!!!!!')

    #user input
    user_input=int(input('\nEnter any number: '))

    #displaying cubes of a given input 
    print()
    for i in range(1,user_input+1):
        if i==user_input:
            cube=i**3
            print('The cube of',i,'is',cube)
    print()
    print('Thankyou!..')
    print('_'*20)



def  mind_game():
    import time
    import random 

    #mind game 
    print("Welcome to the Mind Game!!!!!")
    time.sleep(1)

    #greet user 
    name=input("\nEnter your name: ")
    print('Hello ',name.title())
    time.sleep(2)

    #guess the number in your mind
    print("\n1) Choose any number (1-50000) in your mind .")
    time.sleep(5)

    #double the number
    print("\n2) Now double the number you have choosen.")
    time.sleep(5)

    #adding 4 to the number
    print("\n3) Now add this 4 number accordingly to your answer....... ")
    n1=random.randrange(1,50)
    print('Add: ', n1)
    n2=random.randrange(1,50)
    print('Add: ', n2)
    n3=random.randrange(1,50)
    print('Add: ', n3)
    n4=random.randrange(1,50)
    print('Add: ', n4)
    total=(n1+n2+n3+n4)/2
    print()

    #half 
    print("\n4) Now divide the answer by 2 to get the half of the number.")
    time.sleep(5)

    # - guessed number
    print("\n5) Alright now minus the number that you had guessed earlier.")
    time.sleep(5)

    #final answer
    print("\nUMMMM..........")
    time.sleep(2)
    print('I am guessing that your answer is........')
    time.sleep(2)
    print(total)



def countdown_timer():
    #importing library 
    from countdown import countdown
    #declaring  two variables to display minutes and seconds
    m=int(input('Enter miniutes: '))
    s=int(input("Enter seconds: "))
    # Create a countdown of 1 minute and 10 seconds
    countdown(m,s)
    print("Time's up!!!!")

def table():
    #taking user input
    user_input=int(input('Enter the number: '))

    #table using while loop
    t=1
    while t<11:
        print(user_input,'x',t,'=',t*user_input)
        t=t+1


def dice_game():
    import random

    while True:
        #welcome
        print('\nWelcome to the Dice Game!!!')

        #taking user input
        user_input = int(input('\nEnter your choice (1, 2, 3, 4, 5, 6): '))
        print('\nYou have chosen: ', user_input)

        # Check if the user input is greater than 6
        if int(user_input) > 6:
            print('\nInvalid input. Please choose a number between 1 and 6.')
            continue  # Continue to the next iteration of the loop

        options = [1,2,3,4,5,6]
        computer_choice = random.choice(options)

        print('\nComputer Choice: ', computer_choice)

        if user_input == computer_choice:
            print('\nCongratulations..')
            print("You won!")
        else:
            print("\nComputer Wins this round!!")

        another_round = input("\nDo you want to play again (yes/no): ")
        if another_round.lower() != 'yes':
            print('\nShukriya.....')
            break
    print("_" * 25)



def factorial():
    #welcome 
    print('Welcome to the Factorial Game !!!!!')
    #taking input from the user 
    num=int(input("Enter any number: "))
    fact=1
    #displaying the result by using for-loop
    for i in range(1,num+1):
        fact=fact*i
    print('The factorial of',i,'is',fact)



def fibonacci():
    #Welcome
    print("Welcome to the Fibonaccci Series..")
    #taking user input
    number=int(input('\nEnter the series range: '))
    #declaring n1 and n2 as first term of the series
    n1=0
    n2=1

    #for-loop
    for number in range(0,number):
        if number<=1:
            n3=number
        else:
            n3=n1+n2
            n1=n2
            n2=n3
        print(n3)

    print("The fibonacci series of",number,'is',n3)




def pwd_generator():
    import random
    import string 

    #welcome
    print('Welcome to the Password Generator!!')

    #taking user input
    num=int(input("\nHow many characters of password do you want?: "))

    #declaring one variable to print output in "fgFDD$33*^&%" sequence
    characters = string.ascii_letters + string.digits + string.punctuation

    #initializing an empty string to store the generated password
    password = ''.join(random.choice(characters) for _ in range(num))
    print('\nYour generated password is: ',password)



def calculator():
    #title
    print('---------------CALCULATOR----------------')
    print()

    while True:

        #printing operator list
        print('Choose Any one Operation:- ')
        print('1. Additon(+)')
        print('2. Substraction(-)')
        print('3. Division(/)')
        print('4. Multiplication(*)')
        print()

        #user input 
        user_input=eval(input('Enter your choice (1,2,3,4): ' ))
        print()

        #defining functions
        #addition
        def add(x,y):
            return x + y
        #substraction
        def subs(x,y):
            return x - y
        #multiplication
        def mul(x,y):
            return x * y
        #division
        def div(x,y):
            return x / y
        print()

        #putting if-else condition..
        if user_input==1:
            u1=eval(input('Enter first number: '))
            u2=eval(input('Enter second number: '))
            print()
            print(u1,'+',u2,'=',add(u1,u2))

        elif user_input==2:
            u1=eval(input('Enter first number: '))
            u2=eval(input('Enter second number: '))
            print()
            print(u1,'-',u2,'=',subs(u1,u2))

        elif user_input==3:
            u1=eval(input('Enter first number: '))
            u2=eval(input('Enter second number: '))
            print()
            print(u1,'*',u2,'=',mul(u1,u2))

        elif user_input==4:
            u1=eval(input('Enter first number: '))
            u2=eval(input('Enter second number: '))
            print()
            print(u1,'/',u2,'=',div(u1,u2))

        else:
            print('Invalid input')
            print('Make a choice from above operation')
            print()

        another_calculation=input('Do you want to perform another calculation? (yes/no):  ') 
        if another_calculation.lower() != 'yes':
            print('Shukriya ')
            break

    print('='*20)        



def head_and_tail():
    import random
    #welcome
    print('Welcome to Game!!')
    #user input for choice
    user_input=input('\nEnter your choice(heads or tails): ')
    print("\nYou have chosen: ",user_input.lower())

    if user_input.lower()=='heads' or user_input.lower()=='tails':
        #coin toss
        computer_choice=random.choice(['heads','tails'])
        #display coin result
        print("\nComputer choice: ",computer_choice)
        #compare
        if user_input.lower() == computer_choice:
            print('\nCongratulations!')
            print('You win this round!...')
        else:
            print('\nSorry you have lose..')
    else:
        print('Invalid input')
        print('Please try again....')
        
        
#importing library 
import tkinter

#main window
window=tkinter.Tk()

#window size
window.geometry("810x700")

#title 
window.title('Python Project - Ishita')

#label
label=tkinter.Label(window,text='PYTHON PROJECT',font=('Adobe Caslon Pro Bold',30))
label.place(x=235,y=80)

#label1
label1=tkinter.Label(window,text='Made by - Ishita',font=('Arial CE',10))
label1.place(x=360,y=150)

#button 1
btn1=tkinter.Button(window, text='Budget Game', height=3, width=22, command=budget_game)
btn1.place(x=130,y=250)

#button 2
btn2=tkinter.Button(window, text='Cube Game', height=3, width=22, command=cube_game)
btn2.place(x=330,y=250)

#button 3
btn3=tkinter.Button(window, text='Mind Game', height=3, width=22, command=mind_game)
btn3.place(x=530,y=250)

#button 4
btn4=tkinter.Button(window, text='Countdown Timer', height=3, width=22, command=countdown_timer)
btn4.place(x=130,y=350)

#button 5
btn5=tkinter.Button(window, text='Table', height=3, width=22, command=table)
btn5.place(x=330,y=350)

#button 6
btn6=tkinter.Button(window,text='Dice Game', height=3, width=22, command=dice_game)
btn6.place(x=530,y=350)

#button 7
btn7=tkinter.Button(window, text='Factorial', height=3, width=22, command=factorial)
btn7.place(x=130,y=450)

#button 8
btn8=tkinter.Button(window, text='Fibonacci Series', height=3, width=22, command=fibonacci)
btn8.place(x=330,y=450)

#button 9
btn9=tkinter.Button(window, text='Password Generator', height=3, width=22, command=pwd_generator)
btn9.place(x=530,y=450)

#button 10
btn10=tkinter.Button(window, text='Calculator', height=3, width=22, command=calculator)
btn10.place(x=130,y=550)

#button 11
btn11=tkinter.Button(window, text='Flip It to Win It', height=3, width=22, command=head_and_tail)
btn11.place(x=330,y=550)

#mainloop
window.mainloop()        
```
